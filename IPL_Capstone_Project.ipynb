### Now, Here comes some Basic Questions

1. Which team won the most matches?

```python
df.head()
```

```python
sns.barplot(x = match_win.values , y = match_win.index , palette='viridis')
plt.title("Most match win by team")
```

3. Toss Winner vs Match Winner

```python
toss_match = df[df['toss_winner'] == df['match_winner']]['match_id'].count()

percentage = ((toss_match * 100)/df.shape[0]).round(2)
percentage
```

###  Key Player Performances

1 Most "Player of the Match" Awards

```python
Awards = df['player_of_the_match'].value_counts().head(10)
Awards
```

```python
sns.barplot(y = Awards.index , x = Awards.values , palette='viridis')
```

2 Top Scorers

```python
High = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
High.head(2)
```

```python
High.plot(kind = 'bar')
```

10 Best Bowling Figures

```python
# this code create a new column for extract which player take more wikets in single match.

df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x: x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
```

```python
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers
```

```python
sns.barplot(y = top_bowlers.index , x = top_bowlers.values , palette='rainbow')
```
